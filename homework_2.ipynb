```python
t = np.linspace(0,10, num = 51)
f = np.cos(t)
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t,f, c = 'green', label = 'cosx')
plt.title('Graph f(t)', fontsize = 16, fontweight = 'bold')
plt.xlabel('Values of t')
plt.ylabel('Values of f')
plt.legend(loc = 'lower left', frameon = False)
plt.grid(c = 'lightgrey')
plt.show()
```

```python
x = np.linspace(-3,3, num = 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5 
y4 = np.sin(x) 
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x,y1)
ax1.set_title('Graph y1', fontsize = 10, fontweight = 'bold')
ax1.set_xlim(-5, 5)

ax2.plot(x,y2)
ax2.set_title('Graph y2', fontsize = 10, fontweight = 'bold')

ax3.plot(x,y3)
ax3.set_title('Graph y3', fontsize = 10, fontweight = 'bold')

ax4.plot(x,y4)
ax4.set_title('Graph y4', fontsize = 10, fontweight = 'bold')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```
