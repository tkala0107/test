```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Dataset
read_file = pd.read_excel('data_new.xlsx')
read_file.to_csv(r'data_new.csv')
train=pd.read_csv('data_new.csv')
train=train.drop(['Unnamed: 0'], axis=1)
#train.info()
```

```python
#Dataset test
read_file = pd.read_excel('data_test_new.xlsx')
read_file.to_csv(r'data_test_new.csv')
test=pd.read_csv('data_test_new.csv')
test=test.drop(['Unnamed: 0'], axis=1)
#test.info()
```

```python
train.describe(include = 'all')
```

```python
targets = train['Price']
inputs = train.drop(['Price'], axis = 1)
```

```python
# Import the module for the split
from sklearn.model_selection import train_test_split

# Split the variables with an 80-20 split and some random state

x_train, x_val, y_train, y_val = train_test_split(inputs, targets, test_size=0.2, random_state=365)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 1000, max_depth = 10, random_state = 42)
model.fit(x_train, y_train)
```

```python
y_train_predict = model.predict(x_train)

from sklearn.metrics import r2_score
r2_score(y_train, y_train_predict)
```

```python
importance = model.feature_importances_
feature_importance = pd.DataFrame(importance, index = inputs.columns.values, columns = ['Importance'])
feature_importance.sort_values(by = ['Importance'], ascending = False)
```

```python
train_new = train.drop(['Helthcare_2', 'Rooms', 'Shops_2', 'Ecology_2', 'Ecology_3', 'SquareTest', 'District_cluster'], axis = 1)
```

```python
targets_new = train_new['Price']
inputs_new = train_new.drop(['Price'], axis = 1)
```

```python
x_train_new, x_val_new, y_train_new, y_val_new = train_test_split(inputs_new, targets_new, test_size=0.2, random_state=365)
```

```python
model_new = RandomForestRegressor(n_estimators = 500, max_depth = 10, random_state = 42)
model_new.fit(x_train_new, y_train_new)
```

```python
y_train_new_predict = model_new.predict(x_train_new)

r2_score(y_train_new, y_train_new_predict)
```

```python
y_val_new_predict = model_new.predict(x_val_new)
r2_score(y_val_new, y_val_new_predict)
```

```python
importance = model_new.feature_importances_
feature_importance = pd.DataFrame(importance, index = inputs_new.columns.values, columns = ['Importance'])
feature_importance.sort_values(by = ['Importance'], ascending = False)
```
